```python
import numpy as np
import pandas as pd
import sklearn as sk
import matplotlib.pyplot as plt
from sklearn import linear_model
import sklearn.metrics as sk
```

```python
df=pd.read_csv('/content/drive/MyDrive/dataset/boston.csv')
print(df)
```

```python
y=df['medv'].values
X=df['rm'].values
print(X,y)
```

```python
print(X.shape,y.shape)
```

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)
print(X.shape,y.shape)
```

```python
plt.scatter(X,y)
plt.xlabel('no of rooms')
plt.ylabel('house rate')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
#overfitted model
y_pred = reg.predict(X)
plt.scatter(X,y, color='red')
plt.plot(X,y,color='black',linewidth=1)
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,y_pred,color='black',linewidth=3)
plt.show()
```

```python
print(sk.mean_absolute_error(X,y))
print(sk.r2_score(y,y_pred))
```
