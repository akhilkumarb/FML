# ACTIVATION FUNTION :--> SIGMOID

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))
```

```python
x = np.linspace(-5, 5)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function : --> Sigmoid')
plt.show()
```

# ACTIVATION FUNCTION : --> ReLU

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RELU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)
  return x1
```

```python
x = np.linspace(-5, 5)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function : --> ReLU')
plt.show()
```

# ACTIVATION FUNCTION : --> SOFTMAX

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-5, 5)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function : --> Softmax')
plt.show()
```
