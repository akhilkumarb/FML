```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/dataset/tested.csv')
```

```python
data.info()
```

```python
print(data['Age'].isnull().sum())
print(data['Age'].mean(),data['Age'].median(),data['Age'].std())
```

```python
sns.boxplot(data['Age'])
```

```python
data['Age']=data['Age'].fillna(40)
```

```python
sns.boxplot(data['Age'])
```

```python
x=np.array(data['Age'])
x=x.reshape(-1,1)
```

```python
sca=StandardScaler()
```

```python
y=sca.fit_transform(x)
print(y.mean(),y.std(),y.min(),y.max())
```
